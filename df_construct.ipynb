```python
books = pd.read_csv('/Users/javm/Desktop/Projects/Sistema-Recomendacion-Libros/libros.csv')
books.dropna(inplace = True)
books
```

```python
def clean_authors(text):
    if not text:
        return ''
    pattern = r'[¡!¿?.,:;()\-—«»“”‘’\[\]{}\/\'\"\d]'
    text = re.sub(pattern, '', text)
    text = text.replace('@','')
    return text
```

```python
books['authors'] = books['authors'].apply(clean_authors)
books
```

```python
def clean_blurb(text):
    text = text.replace('Overview\nNotes From Your Bookseller\n', '')
    text = text.replace('Overview\n', '')
    text = text.replace('\n', ' ')
    return text
```

```python
books['blurb'] = books['blurb'].apply(clean_blurb)
books
```

```python
books.to_csv('books.csv', index = False )
```
