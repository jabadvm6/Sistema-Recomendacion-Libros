```python
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

import requests

from bs4 import BeautifulSoup
import time
```

```python

df = pd.DataFrame(columns=['Titulo','Autor']) # creates master dataframe 

# Chromedriver Path
PATH = "/Users/javm/Desktop/chromedriver_mac_arm64/chromedriver"

# this creates/opens Chrome browser
driver = webdriver.Chrome(PATH)

for listings in range(1,2):
    page_num = str(listings)
    url = 'https://www.barnesandnoble.com/b/books/_/N-1z13w9sZ1sZ29Z8q8?Nrpp=20&page=' + page_num
    driver.get(url)
    
    titles = driver.find_elements(By.XPATH,'//div[@class="product-shelf-title product-info-title pt-xs"]') 
    authors = driver.find_elements(By.XPATH,'//div[@class="product-shelf-author pt-0 mt-1"]')

    title_list = []
    for t in range(len(titles)):
        title_list.append(titles[t].text)
    
    author_list = []
    for a in range(len(authors)):
        author_list.append(authors[a].text)
    
    data_tuples = list(zip(title_list[0:],author_list[0:])) # list of each book titlep and author paired together
    book_df = pd.DataFrame(data_tuples, columns=['Titulo','Autor']) # creates dataframe of each tuple in list
    #book_df['Year'] = yr # adds season beginning year to each dataframe
    df = df.append(book_df) # appends to master dataframe
    
driver.close()
```
